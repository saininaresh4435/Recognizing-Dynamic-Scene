```python
#key_frame and key_segment selection
#descrepiton of key_frame_segment_dict selection
#key_frame_segment_dict is a dictionary whose keys are number of videos.
#key_frame_segment_dict={num_of_videos:{num_of_clusters:[],[]-----[]}}
key_frame_segment_dict={}
for i in range(10):
        key_frame_segment_dict[i]={}
for num_of_video in range(10):
    img_data = np.array(img_data_dict_vgg[num_of_video])
    img_data = img_data.astype('float32')
    print (img_data.shape)
    img_data_scaled=img_data.reshape(img_data.shape[0],img_data.shape[2])
    print(img_data_scaled.shape)
    num_of_img=img_data_scaled[0]
    # Import the `cluster` module
    from sklearn import cluster
    from scipy.spatial.distance import euclidean
    import collections
    n_clusters=10
    # Create the KMeans model
    clf = cluster.KMeans(init='k-means++', n_clusters=10,precompute_distances=True, random_state=None)

    # Fit the training data `X_train`to the model
    k=clf.fit(img_data_scaled)
    closest_cluster=k.predict(img_data_scaled)
    cluster_centroid=k.cluster_centers_
    clustering_dict={}
    for i in range(n_clusters):
        clustering_dict[i]=[]
    i=0
    for which_cluster in closest_cluster:
        clustering_dict[which_cluster].append(img_data_dict[num_of_video][i])
        i=i+1
    clustering_dict1={}
    for i in range(n_clusters):
        clustering_dict1[i]=[]
    i=0
    for which_cluster in closest_cluster:
        clustering_dict1[which_cluster].append(img_data_scaled[i])
        i=i+1
    dicta={}
    for i in range(n_clusters):
        dicta[i]={}
    for i in range(n_clusters):
        n=np.array(clustering_dict[i]).shape[0]
        for j in range(n):
            dicta[i][euclidean(cluster_centroid[i],clustering_dict1[i][j])]=clustering_dict[i][j]
    for i in range(n_clusters):
        key_frame_segment_dict[num_of_video][i]=[]
        od = collections.OrderedDict(sorted(dicta[i].items()))
        j=0;
        for k,v in od.items():
            if(j>9):
                break
            key_frame_segment_dict[num_of_video][i].append(v)
            j=j+1  
print(np.array(key_frame_segment_dict[0][0][0]).shape)
```

```python
#feature description
#Sconv_dict={num_of_video:{num_of_cluster:[196*512]}}
#Tconv_dict={num_of_video:{num_of_cluster:[196*512]}}
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
from keras.models import Model
import numpy as np

base_model = VGG16(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.get_layer('block5_conv3').output)

Sconv_dict={}
Tconv_dict={}
for i in range(10):
    Sconv_dict[i]={}
    Tconv_dict[i]={}
for num_of_video in range(10):
    for num_of_cluster in range(n_clusters):
        #for Sconv
        key_frame=key_frame_segment_dict[num_of_video][num_of_cluster][0]
        key_frame_scaled=key_frame.reshape(img_rows,img_cols,num_channel)
        img=np.expand_dims(key_frame_scaled,axis=0)
        Sconv_dict[num_of_video][num_of_cluster]=[]
        features = model.predict(img)
        features = np.array(features).reshape(196,512)
        Sconv_dict[num_of_video][num_of_cluster].append(features)
        #for Tconv
        C=[]
        num=np.array(key_frame_segment_dict[num_of_video][num_of_cluster]).shape[0]
        tau=8
        if(num>=tau):
            for i in range(tau):
                img=key_frame_segment_dict[num_of_video][num_of_cluster][i]
                img=key_frame.reshape(img_rows,img_cols,num_channel)
                img=np.expand_dims(key_frame_scaled,axis=0)
                features = model.predict(img)
                features = np.array(features).reshape(196,512)
                C.append(features)
        else:
            for j in range(num):
                img=key_frame_segment_dict[num_of_video][num_of_cluster][j]
                img=key_frame.reshape(img_rows,img_cols,num_channel)
                img=np.expand_dims(key_frame_scaled,axis=0)
                features = model.predict(img)
                features = np.array(features).reshape(196,512)
                C.append(features)
        mu=np.zeros((196,512))        
        if(num>=tau):
            for i in range(tau):
                mu=mu+np.array(C[i])
            mu=mu/(tau+1)
        else:
            for j in range(num):
                mu=mu+np.array(C[j])
            mu=mu/tau+1
        t=np.zeros((196,512))    
        if(num>=tau):
            for i in range(tau):
                t=t+(np.array(C[i])-mu)**2
            t=t/(tau+1)
        else:
            for j in range(num):
                  t=t+(np.array(C[j])-mu)**2
            t=t/(tau+1)
        Tconv_dict[num_of_video][num_of_cluster]=[]
        Tconv_dict[num_of_video][num_of_cluster].append(t)
```
